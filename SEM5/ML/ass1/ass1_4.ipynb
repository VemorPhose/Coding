### Write a program to create a synthetic dataset with two variables (having non-linear relation). Create a scatter plot to visualize the relationship between X and Y. Report and interpret the value of covariance and the Pearsonâ€™s correlation coefficient between them.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

X_nl = np.linspace(-5, 5, 100)
Y_nl = X_nl**2 + np.random.normal(0, 5, X_nl.shape)
```

```python
np.savetxt('1_4_non_linear_data.csv', np.column_stack((X_nl, Y_nl)), delimiter=',', header='X_nl,Y_nl', fmt='%.6f')
```

```python
plt.figure(figsize=(7, 5))
plt.scatter(X_nl, Y_nl, color='purple')
plt.title('Non-linear Relationship (Quadratic)')
plt.xlabel('X')
plt.ylabel('Y')

plt.savefig('1_4_non_linear_scatter.png')
plt.show()
```

```python
cov_nl = np.cov(X_nl, Y_nl)[0, 1]
pearson_nl = np.corrcoef(X_nl, Y_nl)[0, 1]

print(f"Covariance: {cov_nl:.2f}")
print(f"Pearson's r: {pearson_nl:.2f}")

print("\nCovariance may be non-zero, but Pearson's r will be close to 0, indicating no linear relationship despite a clear non-linear pattern.")
```
