### Write a program to create two different synthetic datasets, each having two variables (one dataset having negative linear correlation and the other dataset having positive linear correlation between variables). Create a scatter plot to visualize the relationship between X and Y. Report and interpret the value of covariance and Pearsonâ€™s correlation coefficient between them.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

X_pos = np.linspace(1, 10, 10)
y_pos = 2 * X_pos + np.random.normal(0, 1, X_pos.shape)

X_neg = np.linspace(1, 10, 10)
y_neg = -2 * X_neg + np.random.normal(0, 1, X_neg.shape)
```

```python
np.savetxt('1_2_correlation_data_pos.csv', np.column_stack((X_pos, y_pos)), delimiter=',', header='X_pos,y_pos', fmt='%.6f')
np.savetxt('1_2_correlation_data_neg.csv', np.column_stack((X_neg, y_neg)), delimiter=',', header='X_neg,y_neg', fmt='%.6f')
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(X_pos, y_pos, color='green')
plt.title('Positive Linear Correlation')
plt.xlabel('X')
plt.ylabel('Y')

plt.subplot(1, 2, 2)
plt.scatter(X_neg, y_neg, color='red')
plt.title('Negative Linear Correlation')
plt.xlabel('X')
plt.ylabel('Y')

plt.tight_layout()
plt.savefig('1_2_correlation_scatter.png')
plt.show()
```

```python
cov_pos = np.cov(X_pos, y_pos)[0, 1]
cov_neg = np.cov(X_neg, y_neg)[0, 1]

pearson_pos = np.corrcoef(X_pos, y_pos)[0, 1]
pearson_neg = np.corrcoef(X_neg, y_neg)[0, 1]

print(f"Positive correlation dataset:")
print(f"Covariance: {cov_pos:.2f}")
print(f"Pearson's r: {pearson_pos:.2f}")

print(f"\nNegative correlation dataset:")
print(f"Covariance: {cov_neg:.2f}")
print(f"Pearson's r: {pearson_neg:.2f}")

print("\nCovariance and Pearson's r are positive for the first dataset, indicating a positive linear relationship.")
print("Covariance and Pearson's r are negative for the second dataset, indicating a negative linear relationship.")
```
