### Write a program to create two different synthetic datasets, each having two variables (one dataset having high variance and the other dataset having low variance). Create a scatter plot to visualize the relationship between X and Y. Report and interpret the value of covariance and the Pearsonâ€™s correlation coefficient between them.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

X_low = np.linspace(1, 10, 20)
y_low = 3 * X_low + np.random.normal(0, 1, X_low.shape)

X_high = np.linspace(1, 10, 20)
y_high = 3 * X_high + np.random.normal(0, 10, X_high.shape)
```

```python
np.savetxt('1_3_variance_data_low.csv', np.column_stack((X_low, y_low)), delimiter=',', header='X_low,y_low', fmt='%.6f')
np.savetxt('1_3_variance_data_high.csv', np.column_stack((X_high, y_high)), delimiter=',', header='X_high,y_high', fmt='%.6f')
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(X_low, y_low, color='green')
plt.title('Low Variance Dataset')
plt.xlabel('X')
plt.ylabel('Y')

plt.subplot(1, 2, 2)
plt.scatter(X_high, y_high, color='red')
plt.title('High Variance Dataset')
plt.xlabel('X')
plt.ylabel('Y')

plt.tight_layout()
plt.savefig('1_3_variance_scatter.png')
plt.show()
```

```python
cov_pos = np.cov(X_low, y_low)[0, 1]
cov_neg = np.cov(X_high, y_high)[0, 1]

pearson_pos = np.corrcoef(X_low, y_low)[0, 1]
pearson_neg = np.corrcoef(X_high, y_high)[0, 1]

print(f"Low variance dataset:")
print(f"Covariance: {cov_pos:.2f}")
print(f"Pearson's r: {pearson_pos:.2f}")

print(f"\nHigh variance dataset:")
print(f"Covariance: {cov_neg:.2f}")
print(f"Pearson's r: {pearson_neg:.2f}")

print("\nThe low variance dataset has a higher Pearson's r, indicating a stronger linear relationship.")
print("The high variance dataset has a lower Pearson's r, showing that increased noise weakens the linear relationship, even if covariance is high.")
```
